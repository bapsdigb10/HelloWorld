```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
url=r"C:\Users\LENOVO\Downloads\Weather_Data.csv"
df=pd.read_csv(url)
df.head()
```

```python
df.describe
```

```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

#### TASK 1: Visualize the relationship between Flight Number and Launch Site

```python
sns.catplot(x="FlightNumber",y="LaunchSite",data=df,hue="Class")
plt.title("Relationship between Flight Number and Launch Site")
plt.xlabel("Flight Number")
plt.ylabel("Launch Site")
plt.show()
```

```python
sns.catplot(x="PayloadMass",y="LaunchSite",data=df,hue="Class")
plt.title("Relationship between Payload Mass and Launch Site")
plt.xlabel("Payload Mass")
plt.ylabel("Launch Site")
plt.show()
```

#### TASK 3: Visualize the relationship between success rate of each orbit type

```python
success_rate_by_orbit = df.groupby('Orbit')['Class'].mean()*100
success_rate_by_orbit.plot(kind="bar",color="blue")
plt.title("Relationship between Orbit Type and Success Rate")
plt.xlabel("Orbit Type")
plt.ylabel("Success Rate")
plt.plot()
```

#### TASK 6: Visualize the launch success yearly trend

```python
# A function to Extract years from the date 
year=[]
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year
Extract_year()
df['Date'] = year
df.head()
```

```python
success_rate_by_year = df.groupby('Date')['Class'].mean()
success_rate_by_year.plot(kind='line', color='green', marker='o',linestyle="dashed")
plt.title('Launch Success Rate Yearly Trend', fontsize=16)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Average Success Rate', fontsize=12)
plt.show()
```
