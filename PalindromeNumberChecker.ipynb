```python
##PALINDROME CHECKER
rev=0
val=int(input("ENTER THE THREE DIGIT NUMBER"))
print(val)
temp=val
while val > 0:
    num=val % 10
    rev=(rev*10)+num
    val=val // 10
if temp==rev:
    print("PALINDROME NUMBER")
else:
    print("NOT PALINDROME NUMBER")
```
